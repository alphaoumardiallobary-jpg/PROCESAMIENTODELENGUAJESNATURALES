# 1- Analisis de sentimientos basado en reglas

### Textblob 

```python
from textblob import TextBlob
```

```python
sentence_1 = "I loved the movie, it was amazing."
sentence_2 = "The film was fantastic and very entertaining."
sentence_3 = "I had a great time watching the movie."
sentence_4 = "The movie was funny and enjoyable."
sentence_5 = "It was a wonderful movie with a great story."
sentence_6 = "The actors did an excellent job."
sentence_7 = "The film was exciting and very well made."
sentence_8 = "The movie was boring and too long."
sentence_9 = "I did not enjoy the movie at all."
sentence_10 = "The film was terrible and disappointing."
sentence_11 = "The acting was bad and the story was weak."
sentence_12 = "It was a very dull movie."
sentence_13 = "The movie was awful and poorly made."
sentence_14 = "I regret watching that movie."
sentence_15 = "I went to see a movie yesterday."
sentence_16 = "We watched a film at the cinema."
sentence_17 = "The movie started at eight o'clock."
sentence_18 = "I bought popcorn before the movie."
sentence_19 = "The cinema was full of people."
sentence_20 = "We sat in the middle row."
```

```python
print(sentence_1)
sentiment_score = TextBlob(sentence_1)
print(sentiment_score.sentiment.polarity)
```

```python
print(sentence_2)
sentiment_score_2 = TextBlob(sentence_2)
print(sentiment_score_2.polarity)
```

```python
print(sentence_3)
sentiment_score_3 = TextBlob(sentence_3)
print(sentiment_score_3.polarity)
```

```python
print(sentence_9)
sentiment_score = TextBlob(sentence_9)
print(sentiment_score.sentiment.polarity)
```

```python
print(sentence_10)
sentiment_score = TextBlob(sentence_10)
print(sentiment_score.sentiment.polarity)
```

```python
print(sentence_11)
sentiment_score = TextBlob(sentence_11)
print(sentiment_score.sentiment.polarity)
```
