# Reconocimiento de entidades nombradas

```python
import spacy
from spacy import displacy
from spacy import tokenizer
import re
nlp = spacy.load('en_core_web_sm')
```

```python
text = ("Amazon was founded on July 5, 1994 by Jeff Bezos in the city of Seattle in the state of Washington, United States. The company started as an online bookstore, but over time it expanded to sell a wide variety of products, including electronics, clothing, toys, and many other items. As e-commerce grew, Amazon became one of the largest companies in the world.Over the years, Amazon developed new services and products such as Amazon Prime, a subscription service that offers fast shipping and access to digital content. The company also created Amazon Web Services, a cloud computing platform used by thousands of businesses around the world. Today, Amazon is one of the most influential technology companies globally.")
print(text)
```

```python
spacy_doc = nlp(text)
```

```python
for word in spacy_doc.ents:
    print(word.text, word.label_)
```

```python
displacy.render(spacy_doc,style="ent",jupyter=True)
```

```python
text_clean = re.sub(r'[^\w\s]', '', text).lower()
print(text_clean)
```

```python
spacy_doc_clean = nlp(text_clean)
```

```python
for word in spacy_doc_clean.ents:
    print(word.text, word.label_)
```

```python
displacy.render(spacy_doc_clean,style="ent",jupyter=True)
```
