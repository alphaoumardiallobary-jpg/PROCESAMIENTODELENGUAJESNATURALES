# Bolsa de palabras

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = ["I really enjoyed reading this book because the story was very interesting and the characters were well developed and easy to understand",
"This movie was extremely boring and too long, and I felt that the story moved very slowly and did not keep my attention",
"The novel had an exciting plot with many surprising moments and the author did a great job describing the characters and their emotions",
"I did not like this book very much because the writing style was confusing and the story did not make much sense to me",
"The movie was amazing and entertaining from the beginning to the end, and the actors gave very strong performances",
"I thought the story was interesting at first, but later the plot became repetitive and the ending was disappointing",
"really enjoyed this story because it was funny, emotional, and very easy to read, which made the experience very pleasant",
"This book had a good idea and an interesting concept, but the author did not develop the characters enough to make the story memorable",
"The film was entertaining and exciting, with beautiful scenes and a strong storyline that kept me interested the whole time",
"I regret watching this movie because the acting was weak, the dialogue was unrealistic, and the story was not engaging at all."]
```

```python
countvec = CountVectorizer()
```

```python
countvec_fit = countvec.fit_transform(data)
```

```python
bag_of_words = pd.DataFrame(countvec_fit.toarray(), columns = countvec.get_feature_names_out())
print(bag_of_words)
```

# TF-IDF

```python
tfidfvec = TfidfVectorizer()
```

```python
tfidfvec_fit = tfidfvec.fit_transform(data)
```

```python
tfidf_bag = pd.DataFrame(tfidfvec_fit.toarray(), columns = tfidfvec.get_feature_names_out())
print(tfidf_bag)
```
